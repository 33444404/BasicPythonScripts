```python
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
df = pd.read_csv("Files\\Text\\New People List.csv")
```

```python
random_numbers = []
for i in range(1,len(df)+1):
    random_numbers.append(random.randint(1, 1000))

df["Numbers"] = random_numbers

print(df["Numbers"])
```

```python
df["Numbers"].plot()
plt.show
```
